```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("report.csv",delim_whitespace=True)
```

```python
predators = df[df.columns[1]].values
```

```python
preys = df[df.columns[2]].values
```

```python
plt.figure(figsize=(20,15))
plt.plot(preys, label='preys')
plt.plot(predators, label='predators')
plt.legend()
plt.show()
```
